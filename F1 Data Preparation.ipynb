```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
races = pd.read_csv(r"C:\Users\morch\Downloads\ACM Event\races.csv")
drivers = pd.read_csv(r"C:\Users\morch\Downloads\ACM Event\drivers.csv")
lap_times = pd.read_csv(r"C:\Users\morch\Downloads\ACM Event\lap_times.csv")
pit_stops = pd.read_csv(r"C:\Users\morch\Downloads\ACM Event\pit_stops.csv")
constructors = pd.read_csv(r"C:\Users\morch\Downloads\ACM Event\constructors.csv")
results = pd.read_csv(r"C:\Users\morch\Downloads\ACM Event\results.csv")
qualifying = pd.read_csv(r"C:\Users\morch\Downloads\ACM Event\qualifying.csv")
```

```python
lap_times_races = pd.merge(lap_times, races[['raceId', 'year', 'name']], on='raceId')

lap_times_results = pd.merge(lap_times_races, results[['raceId', 'driverId', 'constructorId']], on=['raceId', 'driverId'])

lap_times_constructors = pd.merge(lap_times_results, constructors[['constructorId', 'name']], on='constructorId')

years = [2010, 2020, 2021]
fastest_laps = lap_times_constructors[lap_times_constructors['year'].isin(years)]

fastest_laps_summary = fastest_laps.groupby(['year', 'name_x', 'name_y'])['milliseconds'].min().reset_index()

fastest_laps_summary.columns = ['Year', 'EventName', 'Constructor', 'FastestLapTime']

fastest_laps_summary
```

```python
fastest_laps_summary.to_csv(r'C:\Users\morch\Downloads\ACM Event\fastest_laps_summary.csv', index=False)
```

```python
pit_stops_races = pd.merge(pit_stops, races[['raceId', 'year', 'name']], on='raceId')

pit_stops_results = pd.merge(pit_stops_races, results[['raceId', 'driverId', 'constructorId']], on=['raceId', 'driverId'])

pit_stops_constructors = pd.merge(pit_stops_results, constructors[['constructorId', 'name']], on='constructorId')

years = [2020, 2021, 2022, 2023]
fastest_pit_stops = pit_stops_constructors[pit_stops_constructors['year'].isin(years)]

fastest_pit_stops_summary = fastest_pit_stops.groupby(['year', 'name_x', 'name_y'])['milliseconds'].min().reset_index()

fastest_pit_stops_summary.columns = ['Year', 'EventName', 'Constructor', 'FastestPitStopTime']

fastest_pit_stops_summary
```

```python
fastest_pit_stops_summary.to_csv(r'C:\Users\morch\Downloads\ACM Event\fastest_pit_stops_summary.csv', index=False)
```

```python
constructor_standings = pd.read_csv(r"C:\Users\morch\Downloads\ACM Event\constructor_standings.csv")

constructor_standings_with_year = pd.merge(constructor_standings, races[['raceId', 'year']], on='raceId')

constructors_standings_merged = pd.merge(constructor_standings_with_year, constructors[['constructorId', 'name']], on='constructorId')

teams = ['McLaren', 'Ferrari']
constructors_comparison = constructors_standings_merged[constructors_standings_merged['name'].isin(teams)]

constructors_summary = constructors_comparison.groupby(['year', 'raceId', 'name'])['position'].mean().reset_index()

constructors_summary.columns = ['Year', 'RaceID', 'Constructor', 'AvgPosition']

constructors_summary
```

```python
constructors_summary.to_csv(r'C:\Users\morch\Downloads\ACM Event\constructors_summary_with_raceid.csv', index=False)
```

```python
hamilton_id = drivers[drivers['surname'] == 'Hamilton']['driverId'].values[0]

hamilton_qualifying = qualifying[qualifying['driverId'] == hamilton_id]
hamilton_qualifying = pd.merge(hamilton_qualifying, races[['raceId', 'year']], on='raceId')

years = [2021, 2022, 2023]
hamilton_qualifying_filtered = hamilton_qualifying[hamilton_qualifying['year'].isin(years)]

hamilton_q3_failures = hamilton_qualifying_filtered[hamilton_qualifying_filtered['position'] > 10]

hamilton_q3_failures_summary = hamilton_q3_failures.groupby('year')['position'].count().reset_index()

hamilton_q3_failures_summary.columns = ['Year', 'Q3Failures']

hamilton_q3_failures_summary
```

```python
hamilton_q3_failures_summary.to_csv(r'C:\Users\morch\Downloads\ACM Event\hamilton_q3_failures_summary.csv', index=False)
```
